# ✈️ Flight Delay Analysis — Data Preprocessing

This notebook loads the raw dataset, cleans it, engineers features, and saves the cleaned output.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('Loading dataset...')
df = pd.read_csv('../data/flights_sample_3m.csv', low_memory=False)
print(f'Raw shape: {df.shape}')
df.head()
```

```python
# --- Inspect structure ---
print('=== DATA TYPES ===')
print(df.dtypes)
print('\n=== NULL COUNTS ===')
print(df.isnull().sum())
print('\n=== BASIC STATS ===')
df.describe()
```

```python
# --- Handle missing values ---
# Keep only flights that actually departed (not cancelled)
df = df[df['CANCELLED'] == 0].copy()

# Drop rows where core delay fields are missing
required_cols = ['DEP_DELAY', 'ARR_DELAY', 'DEP_TIME', 'CRS_DEP_TIME', 'DISTANCE']
df = df.dropna(subset=required_cols)

print(f'After dropping NaN rows: {df.shape}')
```

```python
# --- Fix delay values (clip early departures/arrivals to 0) ---
df['DEP_DELAY'] = df['DEP_DELAY'].clip(lower=0)
df['ARR_DELAY'] = df['ARR_DELAY'].clip(lower=0)

# --- Convert FL_DATE to datetime ---
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

# --- Extract temporal features ---
df['MONTH'] = df['FL_DATE'].dt.month
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.dayofweek  # 0=Monday, 6=Sunday

# --- Extract HOUR from CRS_DEP_TIME (integer hhmm format) ---
df['HOUR'] = (df['CRS_DEP_TIME'] // 100).astype(int)
df['HOUR'] = df['HOUR'].clip(0, 23)  # handle midnight edge-case (2400 -> 0)

# --- Time of day category ---
def time_of_day(hour):
    if hour < 6:
        return 'Night'
    elif hour < 12:
        return 'Morning'
    elif hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['TIME_OF_DAY'] = df['HOUR'].apply(time_of_day)

# --- Target variable ---
df['DELAYED'] = (df['DEP_DELAY'] > 15).astype(int)

print('Feature engineering complete.')
print(df[['FL_DATE', 'AIRLINE_CODE', 'ORIGIN', 'DEP_DELAY', 'ARR_DELAY', 'HOUR', 'TIME_OF_DAY', 'DELAYED']].head(10))
```

```python
# --- Summary statistics after cleaning ---
print('=== CLEANED DATASET SUMMARY ===')
print(f'Total flights: {len(df):,}')
print(f'Delayed flights (>15 min): {df["DELAYED"].sum():,} ({df["DELAYED"].mean()*100:.1f}%)')
print(f'Avg departure delay: {df["DEP_DELAY"].mean():.2f} min')
print(f'Avg arrival delay: {df["ARR_DELAY"].mean():.2f} min')
print(f'Airlines: {df["AIRLINE_CODE"].nunique()}')
print(f'Origin airports: {df["ORIGIN"].nunique()}')
print(f'Date range: {df["FL_DATE"].min()} to {df["FL_DATE"].max()}')
```

```python
# --- Save cleaned data ---
output_path = '../data/flights_clean.csv'
df.to_csv(output_path, index=False)
print(f'✅ Cleaned data saved to: {output_path}')
print(f'   Shape: {df.shape}')
```
